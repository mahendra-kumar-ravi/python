### Plotting multiple lines 


```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
```

```python
import numpy as np 
import pandas as pd
```

```python
plt.plot(years, apples,marker="v", color='green')
plt.plot(years,oranges, marker="^")
plt.xlabel("years ")
plt.ylabel("Yield of crop ");
```

```python
plt.plot(years,apples,marker='o', c="green")
plt.plot(years,oranges, marker='x', c='blue');
plt.xlabel("years")
plt.ylabel("Yield")

plt.title("crop yields in kanto");
plt.legend(['Apples','Oranges']);
```

```python
plt.plot(years,apples,marker='s',c='b', ls="-",lw=2, ms=5, mec='y');
plt.plot(years,oranges,marker='o',c='r', ls="--",alpha=.5,lw=2, ms=8, mec='green');

plt.xlabel("year")
plt.ylabel("Yield  ")

plt.title("Crop yield in kanto ")
plt.legend(['Apples', 'Oranges']);
```

```python
# sort hand marker styles 
plt.plot(years,apples, 's-b')
plt.plot(years,oranges, '^--g')

plt.xlabel('Years')
plt.ylabel('Yield')

plt.title("Crop yield in kanto ")
plt.legend(['Apples', 'Oranges']);
```

```python
#  Always write fighure at the top of the plot execution that's why it can be reduced directly 

plt.figure(figsize=(4,4));
plt.plot(years,oranges, 'or')
plt.title("Crop yield in kanto ");
```
